```python
def pie_chart(column):
    def fmt(x):
        return '{:.1f}%\n{:.0f}'.format(x, len(churn_df)*x/100)
    plt.title(column)
    plt.pie(
        churn_df[column].value_counts().values,
        labels=churn_df[column].value_counts().index,
        autopct=fmt)
pie_chart('SubscriptionType')
```

```python
churned = churn_df[churn_df['Churn'] == 1]
stayed = churn_df[churn_df['Churn'] == 0]
```

```python
churn_df['SubscriptionType'].value_counts().index
```

```python
def stacked_bar(column):
       fig, ax = plt.subplots()
       ax.bar(churn_df[column].value_counts().index,
              stayed[column].value_counts(), label='Stayed')
       # Then plot the 'Female' bars on top, starting at the top of the 'Male'
       # bars.
       ax.bar(churn_df[column].value_counts().index,
              churned[column].value_counts(),
              bottom=stayed[column].value_counts(),
              label='Churned')
       ax.set_title(column)
       ax.legend()
stacked_bar(categorical[0])
```
